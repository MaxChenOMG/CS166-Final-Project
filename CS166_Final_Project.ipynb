# Convert the positive MAF file into a sample × metabolite format.
We require three distinct types of data matrices:

1. Abundance Matrix – the baseline feature matrix based on metabolite intensities across samples.

2. Morgan Fingerprint Matrix – structure-based features derived from the molecular structure of each metabolite using extended-connectivity fingerprints (ECFP).

3. ChemBERTa Embedding Matrix – deep learning-based structural representations obtained from the ChemBERTa pretrained model.

First read the MAF (positive mode)

```python
import pandas as pd

pos = pd.read_csv("m_MTBLS8920_LC-MS_positive_hilic_metabolite_profiling_v2_maf.tsv", sep="\t")
print(pos.shape)
pos.head()
```

Log transformation reduces skewness, compresses outliers, and helps approximate a normal distribution.
Standardization (z-scoring) ensures all features contribute equally by scaling them to a common range, which is crucial for many distance-based and gradient-based learning algorithms.

```python
# 1. Split X, y
X = abundance_ML.drop(columns=["sample", "label"])
y = abundance_ML["label"]

# 2. Log transform (safe, deterministic)
import numpy as np
X_log = np.log1p(X)

# 3. Build pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=1000))
])

# 4. Sanity check shapes
X_log.shape, y.shape
```

```python
# =============================
# Abundance-only ML baseline
# =============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    roc_auc_score,
    roc_curve,
    confusion_matrix,
    classification_report
)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# -------------------------------------
# 1. Build X, y
# -------------------------------------
X = abundance_ML.drop(columns=["sample", "label"])
y = abundance_ML["label"]

# -------------------------------------
# 2. Log transform (safe, deterministic)
# -------------------------------------
X_log = np.log1p(X)

# -------------------------------------
# 3. Train-test split
# -------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_log,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# -------------------------------------
# 4. Define models as PIPELINES
# -------------------------------------
models = {
    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=500))
    ]),
    "SVM (RBF kernel)": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", SVC(kernel="rbf", probability=True))
    ]),
    "Random Forest": Pipeline([
        ("scaler", StandardScaler()),   # harmless, keeps structure consistent
        ("clf", RandomForestClassifier(
            n_estimators=200,
            random_state=42
        ))
    ])
}

# -------------------------------------
# 5. Train & evaluate on test set
# -------------------------------------
results = {}

for name, pipe in models.items():
    print("\n==============================")
    print(f"Training: {name}")
    print("==============================")

    # Train (scaler + model fit ONLY on training data)
    pipe.fit(X_train, y_train)

    # Predictions
    y_prob = pipe.predict_proba(X_test)[:, 1]
    y_pred = pipe.predict(X_test)

    auc = roc_auc_score(y_test, y_prob)

    results[name] = {
        "model": pipe,
        "auc": auc,
        "confusion_matrix": confusion_matrix(y_test, y_pred),
        "classification_report": classification_report(y_test, y_pred)
    }

    print(f"AUC: {auc:.4f}")
    print(results[name]["classification_report"])

# -------------------------------------
# 6. Plot ROC curves
# -------------------------------------
plt.figure(figsize=(8, 6))
for name, result in results.items():
    pipe = result["model"]
    y_prob = pipe.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={result['auc']:.3f})")

plt.plot([0, 1], [0, 1], "k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves — Abundance-Only Models")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# -------------------------------------
# 7. Plot confusion matrices
# -------------------------------------
for name, result in results.items():
    cm = result["confusion_matrix"]

    plt.figure(figsize=(4, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix — {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# -------------------------------------
# 8. 5-fold Cross-validation (NO LEAKAGE)
# -------------------------------------
print("\n===== 5-fold Cross Validation =====")
for name, pipe in models.items():
    cv_auc = cross_val_score(
        pipe,
        X_log,
        y,
        cv=5,
        scoring="roc_auc"
    )
    print(f"{name}: mean AUC={cv_auc.mean():.4f}, std={cv_auc.std():.4f}")
```

# ------------------- ChemBERTa -------------------

Same dataframe generation code for sample data and abundance data

```python
import pandas as pd
meta_3 = pd.read_csv("s_MTBLS8920.txt", sep="\t")
meta_clean_3 = meta_3[~meta_3["Sample Name"].str.startswith("QC")].copy()
meta_ml_3 = meta_clean_3[["Sample Name", "Factor Value[Disease]"]].copy()
meta_ml_3 = meta_ml_3.rename(columns={
    "Sample Name": "sample",
    "Factor Value[Disease]": "disease"
})
meta_ml_3["label"] = meta_ml_3["disease"].map({
    "breast cancer": 1,
    "healthy control": 0
})

meta_ml_3
```

```python
pos_3 = pd.read_csv(
    "m_MTBLS8920_LC-MS_positive_hilic_metabolite_profiling_v2_maf.tsv",
    sep="\t"
)
```

# Obtain metabolite ID + SMILES, delete those with empty SMILES

```python
metabolite_smiles_3 = pos_3[["database_identifier", "smiles"]].copy()

metabolite_smiles_3 = metabolite_smiles_3.dropna(subset=["smiles"])
metabolite_smiles_3 = metabolite_smiles_3[metabolite_smiles_3["database_identifier"] != "CHEBI:172744"].reset_index(drop=True)

metabolite_smiles_3


```

```python
!pip install transformers

from transformers import AutoTokenizer, AutoModel
import torch

chemberta_tokenizer_3 = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
chemberta_model_3 = AutoModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
```

We generate ChemBERTa embeddings for each metabolite. As commonly used in recent papers and state-of-the-art (SOTA) methods, each metabolite is mapped to a 768-dimensional vector representation.

```python
def get_chemberta_embedding_3(smiles):
    try:
        encoded_input = chemberta_tokenizer_3(
            smiles,
            padding=True,
            truncation=True,
            return_tensors="pt"
        )
        with torch.no_grad():
            output = chemberta_model_3(**encoded_input)
        # CLS embedding
        emb = output.last_hidden_state[:, 0, :].squeeze().numpy()
        return emb
    except:
        return None
```

# Generate a dictionary mapping each metabolite to its embedding vector.

```python
chemberta_dict_3 = {}

for idx, row in metabolite_smiles_3.iterrows():
    met_id = row["database_identifier"]
    smi = row["smiles"]
    emb = get_chemberta_embedding_3(smi)
    if emb is not None:
        chemberta_dict_3[met_id] = emb
```

# We combine the ChemBERTa embedding (structural features) with the abundance matrix (quantitative features) for each metabolite to form a unified input representation.

```python
import pandas as pd
import numpy as np

abundance_3 = pos_3.set_index("database_identifier").T

print("Abundance matrix shape:", abundance_3.shape)
abundance_3.head()
```

```python
abundance_3 = abundance_3.apply(pd.to_numeric, errors="coerce")
abundance_3 = abundance_3.fillna(0)

# ------------------------------------------
# Step 6.5 — Log-transform (reduce magnitude)
# ------------------------------------------
abundance_3 = np.log1p(abundance_3)

# Keep only metabolites with ChemBERTa embeddings
valid_mets_3 = [m for m in abundance_3.columns if m in chemberta_dict_3]
abundance_3 = abundance_3[valid_mets_3]

# ------------------------------------------
# Step 6.6 — StandardScaler normalization
# ------------------------------------------
from sklearn.preprocessing import StandardScaler

scaler_ab_3 = StandardScaler()
abundance_3_scaled = pd.DataFrame(
    scaler_ab_3.fit_transform(abundance_3),
    index=abundance_3.index,
    columns=abundance_3.columns
)

print("Number of metabolites with ChemBERTa embeddings:", len(valid_mets_3))

# ------------------------------------------
# Step 7 — Compute weighted ChemBERTa sample embedding
# ------------------------------------------
sample_embeddings_3 = {}

for sample in abundance_3_scaled.index:
    emb_vec = np.zeros(768)

    for met in valid_mets_3:
        weight = abundance_3_scaled.loc[sample, met]
        emb = chemberta_dict_3[met]
        emb_vec += weight * emb

    sample_embeddings_3[sample] = emb_vec

# Convert to DataFrame
sample_embed_df_3 = pd.DataFrame.from_dict(sample_embeddings_3, orient="index")
sample_embed_df_3.index.name = "sample"

# Merge with labels
chemberta_ML_3 = sample_embed_df_3.merge(
    meta_ml_3[["sample", "label"]],
    on="sample",
    how="inner"
)

chemberta_ML_3
```

```python
# =============================
# ML: ChemBERTa Embeddings (CORRECT)
# =============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    roc_auc_score,
    roc_curve,
    confusion_matrix,
    classification_report
)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# -------------------------------------
# 1. Prepare data
# -------------------------------------
X_chem = chemberta_ML_3.drop(columns=["sample", "label"]).values.astype(float)
y_chem = chemberta_ML_3["label"].values.astype(int)

# -------------------------------------
# 2. Train-test split FIRST
# -------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_chem,
    y_chem,
    test_size=0.2,
    random_state=42,
    stratify=y_chem
)

# -------------------------------------
# 3. Define models as PIPELINES
# -------------------------------------
models = {
    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=500))
    ]),
    "SVM (RBF kernel)": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", SVC(kernel="rbf", probability=True))
    ]),
    "Random Forest": Pipeline([
        ("scaler", StandardScaler()),   # harmless, keeps interface consistent
        ("clf", RandomForestClassifier(
            n_estimators=200,
            random_state=42
        ))
    ])
}

results = {}

# -------------------------------------
# 4. Train & evaluate (test set)
# -------------------------------------
for name, pipe in models.items():
    print(f"\n==============================")
    print(f"Training: {name}")
    print("==============================")

    pipe.fit(X_train, y_train)

    y_prob = pipe.predict_proba(X_test)[:, 1]
    y_pred = pipe.predict(X_test)

    auc = roc_auc_score(y_test, y_prob)

    results[name] = {
        "model": pipe,
        "auc": auc,
        "confusion_matrix": confusion_matrix(y_test, y_pred),
        "classification_report": classification_report(y_test, y_pred)
    }

    print(f"AUC: {auc:.4f}")
    print(results[name]["classification_report"])

# -------------------------------------
# 5. Plot ROC curves
# -------------------------------------
plt.figure(figsize=(8, 6))
for name, result in results.items():
    pipe = result["model"]
    y_prob = pipe.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={result['auc']:.3f})")

plt.plot([0, 1], [0, 1], "k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves — ChemBERTa Models")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# -------------------------------------
# 6. Confusion matrices
# -------------------------------------
for name, result in results.items():
    cm = result["confusion_matrix"]

    plt.figure(figsize=(4, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix — {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# -------------------------------------
# 7. 5-fold Cross-validation (CORRECT)
# -------------------------------------
print("\n===== 5-fold Cross Validation =====")

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, pipe in models.items():
    cv_auc = cross_val_score(
        pipe,
        X_chem,
        y_chem,
        cv=cv,
        scoring="roc_auc"
    )
    print(f"{name} CV AUC: mean={cv_auc.mean():.4f}, std={cv_auc.std():.4f}")
```
